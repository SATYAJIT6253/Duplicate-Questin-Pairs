```python
def cnt_common_words(row):
    cmn1 = set(map(lambda row : row.lower().strip(),row['question1'].split(" ")))
    cmn2 = set(map(lambda row : row.lower().strip(),row['question1'].split(" ")))
    return len(cmn1 & cmn2)
```

# Token Features

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
stop_words
```

```python
def token_faetures_function(row):
    q1 = row['question1']
    q2 = row['question2']
    
    #this is for handling zero case so that infinite will not come
    SAFE_DIV = 0.0001
    token_features = [0.0]*8
    #spliting the question to tokens
    q1_tokens = q1.split()
    q2_tokens = q2.split()
    
    #check their length
    if len(q1_tokens) == 0 or len(q2_tokens) == 0:
        return token_features
    
    #calculating stopwords
    stop_words = set(stopwords.words('english'))
    q1_stop_words = set([word for word in q1_tokens if word in stop_words])
    q2_stop_words = set([word for word in q2_tokens if word in stop_words])
    
    #calculating non-stop words
    q1_nonstop_words = set([word for word in q1_tokens if word not in stop_words])
    q2_nonstop_words = set([word for word in q2_tokens if word not in stop_words])
    
    #calucalting the common non-stop words
    common_nonstop_words = len(q1_nonstop_words.intersection(q2_nonstop_words))
    
    #calculating the common stop words
    common_stop_words = len(q1_stop_words.intersection(q2_stop_words))
    
    #calculating the common tokens
    common_tokens = len(set(q1_tokens).intersection(set(q2_tokens)))
    
    #setting the token features
    token_features[0] = common_nonstop_words / (min(len(q1_nonstop_words), len(q2_nonstop_words)) + SAFE_DIV)
    token_features[1] = common_nonstop_words / (max(len(q1_nonstop_words), len(q2_nonstop_words)) + SAFE_DIV)
    token_features[2] = common_stop_words / (min(len(q1_stop_words), len(q2_stop_words)) + SAFE_DIV)
    token_features[3] = common_stop_words / (max(len(q1_stop_words), len(q2_stop_words)) + SAFE_DIV)
    token_features[4] = common_tokens / (min(len(q1_tokens), len(q2_tokens)) + SAFE_DIV)
    token_features[5] = common_tokens / (max(len(q1_tokens), len(q2_tokens)) + SAFE_DIV)
    
    #checking the last word equal or not
    token_features[6] = int(q1_tokens[-1] == q2_tokens[-1])
    
    #checking the first word equal or not
    token_features[6] = int(q1_tokens[0] == q2_tokens[0])
    
    return token_features;
```

```python
token_features = df.apply(token_faetures_function, axis=1)

df["cwc_min"]       = list(map(lambda x: x[0], token_features))
df["cwc_max"]       = list(map(lambda x: x[1], token_features))
df["csc_min"]       = list(map(lambda x: x[2], token_features))
df["csc_max"]       = list(map(lambda x: x[3], token_features))
df["ctc_min"]       = list(map(lambda x: x[4], token_features))
df["ctc_max"]       = list(map(lambda x: x[5], token_features))
df["last_word_eq"]  = list(map(lambda x: x[6], token_features))
df["first_word_eq"] = list(map(lambda x: x[7], token_features))
```

```python
import distance

def length_features_function(row):
    
    q1 = row['question1']
    q2 = row['question2']
    
    length_features = [0.0]*3
    
    # spliting the question to tokens
    q1_tokens = q1.split()
    q2_tokens = q2.split()
    
    if len(q1_tokens) == 0 or len(q2_tokens) == 0:
        return length_features
    
    # calculating the abs differnce between two length
    length_features[0] = abs(len(q1_tokens) - len(q2_tokens))
    
    #calculating the Average Token Length of both Questions
    length_features[1] = (len(q1_tokens) + len(q2_tokens))/2
    
    #calculating the longest_substr_ratio
    strs = list(distance.lcsubstrings(q1, q2))
    length_features[2] = len(strs[0]) / (min(len(q1), len(q2)) + 1)
    
    return length_features
```

```python
length_features = df.apply(length_features_function, axis=1)

df['abs_len_diff'] = list(map(lambda x: x[0], length_features))
df['mean_len'] = list(map(lambda x: x[1], length_features))
df['longest_substr_ratio'] = list(map(lambda x: x[2], length_features))
```

```python
from fuzzywuzzy import fuzz

def fetch_fuzzy_features(row):
    
    q1 = row['question1']
    q2 = row['question2']
    
    fuzzy_features = [0.0]*4
    
    # fuzz_ratio
    fuzzy_features[0] = fuzz.QRatio(q1, q2)

    # fuzz_partial_ratio
    fuzzy_features[1] = fuzz.partial_ratio(q1, q2)

    # token_sort_ratio
    fuzzy_features[2] = fuzz.token_sort_ratio(q1, q2)

    # token_set_ratio
    fuzzy_features[3] = fuzz.token_set_ratio(q1, q2)

    return fuzzy_features
```

```python
fuzzy_features = df.apply(fetch_fuzzy_features, axis=1)


df['fuzz_ratio'] = list(map(lambda x: x[0], fuzzy_features))
df['fuzz_partial_ratio'] = list(map(lambda x: x[1], fuzzy_features))
df['token_sort_ratio'] = list(map(lambda x: x[2], fuzzy_features))
df['token_set_ratio'] = list(map(lambda x: x[3], fuzzy_features))
```

# creating the final dataset to train the model


```python
ques_df = df[['question1','question2']]
```

```python
df = df.drop(columns=['id','qid1','qid2','question1','question2'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# merge texts
questions = list(ques_df['question1']) + list(ques_df['question2'])

cv = CountVectorizer(max_features=3000)
q1_arr, q2_arr = np.vsplit(cv.fit_transform(questions).toarray(),2)
```

```python
temp_df1 = pd.DataFrame(q1_arr, index= ques_df.index)
temp_df2 = pd.DataFrame(q2_arr, index= ques_df.index)
temp_df = pd.concat([temp_df1, temp_df2], axis=1)
temp_df
```

```python
final_df = pd.concat([df,temp_df],axis=1)
```

# Training the Model

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(final_df.iloc[:,1:].values,final_df.iloc[:,0].values,test_size=0.2,random_state=2)
```

```python
print(X_train.shape,"---",X_test.shape)
print(Y_train.shape,"---",Y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier()
rf.fit(X_train,Y_train)
Y_pred = rf.predict(X_test)
accuracy_score(Y_test,Y_pred)
```
